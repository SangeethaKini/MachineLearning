## Linear Regression Analysis

##### Linear Regression is a statistical method for determining the output based on one or more inputs. It finds a line that best fits the relationship between the input (independant variable) and the output (dependent variable). 
##### The equation for linear regression is : y = mx + C
###### y = output (dependent variable)
###### m = slope
###### x = input (independent variable)
###### c = intercept
