```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df.head(5)
```

```python
df = df.rename(columns = {'per capita income (US$)': 'percapitaincome'})
df.head(5)
```

```python
plt.scatter(df.year, df.percapitaincome, color='red', marker='+')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']].values, df.percapitaincome)
```

```python
reg.predict([[2020]])
```

```python
plt.scatter(df.year, df.percapitaincome, color='red', marker='+', label = 'Actual')
plt.plot(df.year, reg.predict(df[['year']].values), color='blue', linestyle='-', label ='Regression Line') 
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.legend()
plt.show()
```
